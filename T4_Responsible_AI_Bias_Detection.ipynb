```python
import pandas as pd

gender = []
score = []
selected = []

for i in range(50):
    gender.append("Male")
    score.append(70 + (i % 30))
    selected.append(1)

for i in range(50):
    gender.append("Female")
    score.append(70 + (i % 30))
    selected.append(0)

df = pd.DataFrame({
    "Gender": gender,
    "Score": score,
    "Selected": selected
})

print(df.shape)
df.head()
```

```python
df.to_csv("bias_detection_dataset.csv", index=False)
```

```python
import matplotlib.pyplot as plt

df.groupby('Gender')['Selected'].mean().plot(kind='bar')

plt.xlabel("Gender")
plt.ylabel("Selection Rate")
plt.title("Selection Rate by Gender")

plt.show()
```
